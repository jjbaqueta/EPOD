```python
%%file Main.java

import java.util.ArrayList;
import java.util.Arrays;
import java.util.List;

public class Main {

	/**
	 * Função para imprimir uma lista informado por parâmetro.
	 * @param lista - lista do tipo Pessoa
	 */
	public static void imprimirLista(List<Pessoa> lista) {
		System.out.println(Arrays.toString(lista.toArray()));
	}
	
	public static void main(String[] args) {
		List<Pessoa> minhaLista = new MinhaLista<Pessoa>();
		List<Pessoa> arrayList = new ArrayList<Pessoa>();
		List<Pessoa> linkedList = new ArrayList<Pessoa>();
		
		Pessoa p1 = new Pessoa(1, "Maria");
		Pessoa p2 = new Pessoa(2, "João");
		Pessoa p3 = new Pessoa(3, "Pedro");
		Pessoa p4 = new Pessoa(4, "Lucas");
		Pessoa p5 = new Pessoa(5, "Ana");
		
		System.out.println("Listas após a inserção de elementos:");
		
		// Adicionando pessoas na minha lista
		minhaLista.add(p1);
		minhaLista.add(p2);
		minhaLista.add(p3);
		minhaLista.add(p4);
		minhaLista.add(p5);
		imprimirLista(minhaLista);
		
		// Adicionando pessoas array list
		arrayList.add(p1);
		arrayList.add(p2);
		arrayList.add(p3);
		arrayList.add(p4);
		arrayList.add(p5);
		imprimirLista(arrayList);
		
		// Adicionando pessoas array list
		linkedList.add(p1);
		linkedList.add(p2);
		linkedList.add(p3);
		linkedList.add(p4);
		linkedList.add(p5);
		imprimirLista(linkedList);
		
		System.out.println("\nListas após a remoção de uma pessoa (Pessoa:id(2), nome(João))");
		
		minhaLista.remove(new Pessoa(2, "João"));
		imprimirLista(minhaLista);
		
		arrayList.remove(p2);
		imprimirLista(arrayList);
		
		linkedList.remove(p2);
		imprimirLista(linkedList);
	}
}
```

```python
%%file Pessoa.java

import java.util.Objects;

public class Pessoa {
	private Integer id;
	private String nome;

	public Pessoa(Integer id, String nome) {
		this.id = id;
		this.nome = nome;
	}

	public String getNome() {
		return nome;
	}

	public Integer getID() {
		return id;
	}

	@Override
	public int hashCode() {
		return Objects.hash(id, nome);
	}

	@Override
	public boolean equals(Object obj) {
		if (this == obj)
			return true;
		if (obj == null)
			return false;
		if (getClass() != obj.getClass())
			return false;
		Pessoa other = (Pessoa) obj;
		return Objects.equals(id, other.id) && Objects.equals(nome, other.nome);
	}

	@Override
	public String toString() {
		return "(nome: " + nome + ", ID: " + id + ")";
	}
}
```
