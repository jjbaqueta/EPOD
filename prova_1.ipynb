Classe Main com as implementações dos itens requisitados no exercício 4:  

```python
%%file Main.java

import java.util.Collections;
import java.util.LinkedList;
import java.util.List;

public class Main {
	public static void main(String[] args) {
		List<Filme> filmes = ManipuladorContaCSV.lerArquivo("filmes.csv");

		// 4.A: Ordenar e exibir os filmes por título:
		System.out.println("\nFILMES ORDENADOS POR TÍTULO:");
		Collections.sort(filmes);
		exibirFilmes(filmes);

		// 4.B: Ordenar e exibir os filmes por gênero:
		System.out.println("\nFILMES ORDENADOS POR GÊNERO:");
		ComparadorGenero comparadorGenero = new ComparadorGenero();
		Collections.sort(filmes, comparadorGenero);
		exibirFilmes(filmes);

		ComparadorAvaliacao comparadorAvaliacao = new ComparadorAvaliacao();

		// 4.C: Buscar e exibir o filme com a maior avaliação:
		Filme filmeMaxScore = Collections.max(filmes, comparadorAvaliacao);
		System.out.println("\nFILME COM A MAIOR AVALIAÇÃO: " + filmeMaxScore);

		// 4.D: Buscar e exibir o filme com a menor avaliação:
		Filme filmeMinScore = Collections.min(filmes, comparadorAvaliacao);
		System.out.println("\nFILME COM A MENOR AVALIAÇÃO: " + filmeMinScore);

		// 4.E: Liste todos os filmes cuja receita foi pelo menos 10 vezes 
		// maior que o orçamento inicial
		List<Filme> filmesDobroMais = new LinkedList<Filme>();
		for (Filme filme : filmes) {
			if (filme.getReceita() >= filme.getOrcamento() * 10) {
				filmesDobroMais.add(filme);
			}
		}
		System.out.println("\nFILMES COM RECEITA MAIOR QUE ORÇAMENTO (10x OU MAIS):");
		exibirFilmes(filmesDobroMais);

		// 4.F Listar todos os filmes com classificação "Livre" e com duração máxima de 110 minutos.
		List<Filme> filmesLivre110 = new LinkedList<Filme>();
		for (Filme filme : filmes) {
			if (filme.getClassificao().equals("Livre") && filme.getDuracao() <= 110) {
				filmesLivre110.add(filme);
			}
		}
		System.out.println("\nFILMES COM CLASSIFICAÇÃO LIVRE E COM ATÉ 110 MINUTOS:");
		exibirFilmes(filmesLivre110);

		// 4.G: Contabilizar e exibir a quantidade de filmes por gênero:
		List<String> generos = new LinkedList<String>();
		for (Filme filme : filmes) {
			if (!generos.contains(filme.getGenero())) {
				generos.add(filme.getGenero());
			}
		}
		List<Integer> contagem = new LinkedList<Integer>();
		for (String genero : generos) {
			int contador = 0;
			for (Filme filme : filmes) {
				if (filme.getGenero().equals(genero)) {
					contador++;
				}
			}
			contagem.add(contador);
		}
		System.out.println("\nCONTAGEM DE FILMES POR GÊNERO:");
		for (int i = 0; i < generos.size(); i++) {
			System.out.println("gênero: " + generos.get(i) + ", contagem: " + contagem.get(i));
		}
		System.out.println("TOTAL DE FILMES: " + filmes.size());
	}

	public static void exibirFilmes(List<Filme> filmes) {
		for (Filme filme : filmes) {
			System.out.println(filme);
		}
	}
}
```

Classe Filme definida de acordo com os registros contidos no arquivo "filmes.csv":

```python
%%file Filme.java

public class Filme implements Comparable<Filme>{
	private String titulo;
	private String diretor;
	private Integer anoLancamento;
	private String genero;
	private String classificao;
	private Integer duracao;
	private Double avaliacao;
	private Double orcamento;
	private Double receita;
	
	public Filme(String titulo, String diretor, Integer anoLancamento, String genero, String classificao,
			Integer duracao, Double avaliacao, Double orcamento, Double receita) {
		this.titulo = titulo;
		this.diretor = diretor;
		this.anoLancamento = anoLancamento;
		this.genero = genero;
		this.classificao = classificao;
		this.duracao = duracao;
		this.avaliacao = avaliacao;
		this.orcamento = orcamento;
		this.receita = receita;
	}

	public String getTitulo() {
		return titulo;
	}

	public String getDiretor() {
		return diretor;
	}

	public Integer getAnoLancamento() {
		return anoLancamento;
	}

	public String getGenero() {
		return genero;
	}

	public String getClassificao() {
		return classificao;
	}

	public Integer getDuracao() {
		return duracao;
	}

	public Double getAvaliacao() {
		return avaliacao;
	}

	public Double getOrcamento() {
		return orcamento;
	}

	public Double getReceita() {
		return receita;
	}

	@Override
	public String toString() {
		return "Filme [título:" + titulo + ", diretor:" + diretor + ", lançamento:" + anoLancamento + ", gênero:"
				+ genero + ", classificação:" + classificao + ", duração (min):" + duracao + ", avaliação:" + avaliacao
				+ ", orçamento (em milhões):" + orcamento + ", receita (em milhões):" + receita + "]";
	}

	@Override
	public int compareTo(Filme outroFilme) {
		return this.titulo.compareTo(outroFilme.getTitulo());
	}
}
```

Comparador para o atributo "avaliação":

```python
%%file ComparadorAvaliacao.java

import java.util.Comparator;

public class ComparadorAvaliacao implements Comparator<Filme>{

	@Override
	public int compare(Filme filme1, Filme filme2) {
		return filme1.getAvaliacao().compareTo(filme2.getAvaliacao());
	}
}
```

Comparador para o atributo "gênero":

```python
%%file ComparadorGenero.java

import java.util.Comparator;

public class ComparadorGenero implements Comparator<Filme>{

	@Override
	public int compare(Filme filme1, Filme filme2) {
		return filme1.getGenero().compareTo(filme2.getGenero());
	}
}
```

Classe utilizada para a leitura dos registros contidos no arquivo "filmes.csv"

```python
%%file ManipuladorContaCSV.java

import java.io.BufferedReader;
import java.io.File;
import java.io.FileNotFoundException;
import java.io.FileReader;
import java.io.IOException;
import java.util.ArrayList;
import java.util.List;

public abstract class ManipuladorContaCSV {

	/**
	 * Método para ler o arquivo CSV com os registros das contas
	 * 
	 * @param caminhoArq - caminho para o arquivo CSV de entrada
	 * @return uma lista com objetos do tipo Conta
	 */
	public static List<Filme> lerArquivo(String caminhoArq) {
		List<Filme> filmes = new ArrayList<Filme>();
		String linha = "";
		String csvDivisor = ",";
		BufferedReader br = null;
		boolean primeiraLinha = true;

		try {
			br = new BufferedReader(new FileReader(new File(caminhoArq)));
			while ((linha = br.readLine()) != null) {
				if (primeiraLinha) {
					primeiraLinha = false;
					continue;
				}
				String[] campos = linha.split(csvDivisor);								
				Filme filme = new Filme(campos[0], campos[1], Integer.parseInt(campos[2]), 
						campos[3], campos[4], Integer.parseInt(campos[5]), Double.parseDouble(campos[6]),
						Double.parseDouble(campos[7]), Double.parseDouble(campos[8]));
				filmes.add(filme);
			}
		} catch (FileNotFoundException e) {
			System.out.println(e.getMessage());
			e.printStackTrace();
		} catch (IOException e) {
			System.out.println(e.getMessage());
			e.printStackTrace();
		} finally {
			if (br != null) {
				try {
					br.close();
				} catch (IOException e) {
					System.out.println(e.getMessage());
					e.printStackTrace();
				}
			}
		}
		return filmes;
	}
}
```
